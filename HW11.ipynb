Sympy:

f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30

1. Определить корни

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0


```python
import sympy
import matplotlib as plt
from sympy import symbols, plot_parametric
from sympy.plotting import plot
from sympy.solvers import solve
```

```python
x = symbols('x')
```

```python
plot(((-12)*sympy.sin(sympy.cos(x))*x**4 - 18*(x**3)+5*(x**2) + 10*x - 30), (x, -100, 100))
```

```python
solve(((-12)*sympy.sin(sympy.cos(x))*x**4 - 18*(x**3)+5*(x**2) + 10*x - 30), x)
```

Ответ:
Представленная функция - периодическая. В связи с этим, анализируя график функции и ошибку, выдаваемую при попытке решения из ячейки выше, отметом на постакленные задачи (1-7, исключая 4-ю) будут: бесконечное количество.
